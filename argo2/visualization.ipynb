# Create a Torch.Dataset class for the training dataset

```python
from glob import glob
import pickle
import numpy as np

ROOT_PATH = "../argo2/"

cities = ["austin", "miami", "pittsburgh", "dearborn", "washington-dc", "palo-alto"]
splits = ["train", "test"]

def get_city_trajectories(city="palo-alto", split="train", normalized=False):
    f_in = ROOT_PATH + split + "/" + city + "_inputs"
    inputs = pickle.load(open(f_in, "rb"))
    inputs = np.asarray(inputs)
    
    outputs = None
    
    if split=="train":
        f_out = ROOT_PATH + split + "/" + city + "_outputs"
        outputs = pickle.load(open(f_out, "rb"))
        outputs = np.asarray(outputs)

    return inputs, outputs

class ArgoverseDataset(Dataset):
    """Dataset class for Argoverse"""
    def __init__(self, city: str, split:str, transform=None):
        super(ArgoverseDataset, self).__init__()
        self.transform = transform

        self.inputs, self.outputs = get_city_trajectories(city=city, split=split, normalized=False)

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):

        data = (self.inputs[idx], self.outputs[idx])
            
        if self.transform:
            data = self.transform(data)

        return data

# intialize a dataset
city = 'palo-alto' 
split = 'train'
train_dataset  = ArgoverseDataset(city = city, split = split)
```

# Create a DataLoader class for training

```python
batch_sz = 4  # batch size 
train_loader = DataLoader(train_dataset,batch_size=batch_sz)
```

```python
def grab_coor(city,stage,split="train"):

    fig, axs = plt.subplots(2,len(city)//2, figsize=(10,7), facecolor='w', edgecolor='k', dpi =300 )
    fig.subplots_adjust(hspace = .5, wspace=.1)
    axs = axs.ravel()  
    
    for i,j in zip(city, np.arange(len(city))):
        f_in = ROOT_PATH + split + "/" + i + "_" + stage
        inputs = pickle.load(open(f_in, "rb"))
        inputs = np.asarray(inputs)
        results = {'x':[], 'y':[]}
        for entry in inputs:
            for coor in entry:
                results['x'].append(coor[0])
                results['y'].append(coor[1])
        result = pd.DataFrame(results) 
        result['x_stand'] = (result['x'] - result['x'].mean() ) / result['x'].std()
        result['y_stand'] = (result['y'] - result['y'].mean() ) / result['y'].std()
        
            
        axs[j].xaxis.set_ticks([])
        axs[j].yaxis.set_ticks([])
        axs[j].set_title(i) 
        axs[j].hist2d(result['x_stand'],result['y_stand'],bins= [400,400]) 
    fig.suptitle(stage.upper())  


        
```

```python
def test(city,stage,split="train"):

    fig, axs = plt.subplots(2,len(city)//2, figsize=(10,7), facecolor='w', edgecolor='k', dpi =300 )
    fig.subplots_adjust(hspace = .5, wspace=.1)
    axs = axs.ravel()  
    
    for i,j in zip(city, np.arange(len(city))):
        f_in = ROOT_PATH + split + "/" + i + "_" + stage
        inputs = pickle.load(open(f_in, "rb"))
        inputs = np.asarray(inputs)
        

#         for k in range(inputs.shape[0]):
#             inputs[k] -= inputs[k,0]
            
        
        inputs -= np.repeat(inputs[:, 0, :], inputs.shape[1], axis=0).reshape(inputs.shape[0],inputs.shape[1],2)
        
        results = {'x':[], 'y':[]}
        for entry in inputs:
            for coor in entry:
                results['x'].append(coor[0])
                results['y'].append(coor[1])
        result = pd.DataFrame(results) 

        result['x_stand'] = (result['x'] - result['x'].mean() ) / result['x'].std()
        result['y_stand'] = (result['y'] - result['y'].mean() ) / result['y'].std()
        
            
        axs[j].xaxis.set_ticks([])
        axs[j].yaxis.set_ticks([])
        axs[j].set_title(i) 
        
        axs[j].hist2d(result['x_stand'],result['y_stand'],bins= [30,30])
        
    fig.suptitle(stage.upper())  

```

```python
def grab_all_coor(city,split="train"):

    fig, axs = plt.subplots(2,len(city)//2, figsize=(10,7), facecolor='w', edgecolor='k', dpi =300 )
    fig.subplots_adjust(hspace = .5, wspace=.1)
    axs = axs.ravel()  
    
    for i,j in zip(city, np.arange(len(city))):
        f_in = ROOT_PATH + split + "/" + i + "_" + 'inputs'
        inputs = pickle.load(open(f_in, "rb"))
        inputs = np.asarray(inputs)
        f_out = ROOT_PATH + split + "/" + i + "_" + 'outputs'
        outputs = pickle.load(open(f_out, "rb"))
        outputs = np.asarray(outputs)
        all = np.concatenate((inputs, outputs), axis=1)
        results = {'x':[], 'y':[]}
        for entry in all:
            for coor in entry:
                results['x'].append(coor[0])
                results['y'].append(coor[1])
        result = pd.DataFrame(results) 
        result['x_stand'] = (result['x'] - result['x'].mean() ) / result['x'].std()
        result['y_stand'] = (result['y'] - result['y'].mean() ) / result['y'].std()
        
        axs[j].xaxis.set_ticks([])
        axs[j].yaxis.set_ticks([])
        axs[j].set_title(i) 
        axs[j].hist2d(result['x_stand'],result['y_stand'],bins= [400,400]) 
    fig.suptitle('all'.upper()) 
```

```python
def grab_all_city_norm(city,stage,split="train"):

    fig, axs = plt.subplots(1,1, figsize=(10,7), facecolor='w', edgecolor='k', dpi =300 )
    fig.subplots_adjust(hspace = .5, wspace=.1)
  
    f_in = ROOT_PATH + split + "/" + city[0] + "_" + stage
    all = pickle.load(open(f_in, "rb"))
    all = np.asarray(all)

    for i,j in zip(city[1:], np.arange(len(city))):
        f_in = ROOT_PATH + split + "/" + i + "_" + stage
        inputs = pickle.load(open(f_in, "rb"))
        inputs = np.asarray(inputs)

        all = np.vstack((all, inputs))

    results = {'x':[], 'y':[]}
    for entry in all:
        for coor in entry:
            results['x'].append(coor[0])
            results['y'].append(coor[1])
    result = pd.DataFrame(results) 

    result['x_stand'] = (result['x'] - result['x'].mean() ) / result['x'].std()
    result['y_stand'] = (result['y'] - result['y'].mean() ) / result['y'].std()
    
    axs.xaxis.set_ticks([])
    axs.yaxis.set_ticks([])
    axs.hist2d(result['x_stand'],result['y_stand'],bins= [500,500]) 
    axs.set_title("ALL CITY " + stage.upper())
```
